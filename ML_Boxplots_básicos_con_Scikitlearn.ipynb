# Ejemplo Boxplots en Dataset Iris

Este notebook demuestra cómo realizar visualizaciones individuales de variables numéricas en el dataset **Iris** utilizando boxplots. Estos gráficos permiten detectar valores atípicos, rangos de variabilidad y diferencias, en este caso las características de flores del género *Iris*.


```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from math import ceil

# Cargar datos
iris = load_iris(as_frame=True)
df = iris.frame

# En este caso, load_iris es una función de sklearn que devuelve
# Mostrar primeras filas
df.head()
```

## Opción 1: mostrar boxplots de todas las variables numéricas del dataframe

```python

# En caso de que el dataframe contenga columnas numéricas y categóricas,
# seleccionamos solo las numéricas (si son todas numércias no hay problema)
columnas_numericas = df.select_dtypes(include='number').drop(columns=['target']).columns

# Graficar boxplots
for col in columnas_numericas:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f'Boxplot de {col}')
    plt.xlabel(col)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

## Opción 2: Boxplots de solo algunas variables (indicando los nombres de las columnas)

```python
# Graficar boxplots de algunas variables en particular
# en este caso, indicar sobre qué índices de columnas vamos a iterar
# Ejemplo: ver los boxplots de "sepal length (cm)" y"sepal width (cm)" únicamente
lista_de_variabes = ["sepal length (cm)", 	"sepal width (cm)"]
for col in lista_de_variabes:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col], color='lightblue')
    plt.title(f'Boxplot de {col}')
    plt.xlabel(col)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

## ## Opción 3: Boxplots múltiples en formato cuadriculado)

```python

# Columnas numéricas del DataFrame
columnas_numericas = df.select_dtypes(include='number').drop(columns=['target']).columns
n = len(columnas_numericas)

# Configurar el número de columnas y filas
cols = 2  # si se desean 3 o más columnas modificar aquí
rows = ceil(n / cols)

# Crear subplots
fig, axes = plt.subplots(rows, cols, figsize=(cols * 5, rows * 4))
axes = axes.flatten()

# Dibujar boxplots en cada subplot
for i, col in enumerate(columnas_numericas):
    sns.boxplot(x=df[col], ax=axes[i], color='skyblue')
    axes[i].set_title(f'Boxplot de {col}')
    axes[i].set_xlabel(col)
    axes[i].grid(True)

# Ocultar ejes vacíos si hay
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
