### Tutorial sobre palabras vacías

```python
!python -m spacy download en_core_web_sm
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp("We just opened our wings, the flying part is coming soon")

for token in doc:
    if token.is_stop:
        print(token)
```

```python
def preprocess(text):
    doc = nlp(text)

    no_stop_words = [token.text for token in doc if not token.is_stop]
    return " ".join(no_stop_words)
```

```python
preprocess("Musk wants time to prepare for a trial over his")
```

```python
preprocess("The other is not other but your divine brother")
```

### Eliminar palabras vacías de una columna de texto en un DataFrame de pandas

El conjunto de datos se descarga desde: https://www.kaggle.com/datasets/jbencina/department-of-justice-20092018-press-releases
Contiene comunicados de prensa de distintos casos judiciales del Departamento de Justicia (DOJ). Los comunicados incluyen información como los resultados de casos penales, acciones destacadas contra delincuentes, u otras actualizaciones sobre la administración actual.

```python
import pandas as pd

df = pd.read_json("https://raw.githubusercontent.com/codebasics/nlp-tutorials/refs/heads/main/10_stop_words/doj_press.json",lines=True)

df.shape
```

```python
df.head(5)
```

Filtrar aquellas filas que no tienen temas asociados al caso (columna **topics**)

```python
df = df[df["topics"].str.len() != 0]
#variante con df.query df = df.query("topics != ''")
df.head()
```

```python
df.shape
```

```python
df =df.head(100)
df.shape
```

```python
df["contents_new"] = df.contents.apply(preprocess)
```
