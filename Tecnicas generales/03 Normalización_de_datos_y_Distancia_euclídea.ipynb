# Normalización de datos - Distancia euclídea

## **Ejemplo: Efecto de Diferentes Magnitudes en la Distancia de los Datos**  

Imaginemos que tenemos un conjunto de datos con tres personas donde se registran dos características: **altura (en metros)** y **peso (en kilogramos)**.

| Persona | Altura (m) | Peso (kg) |
|---------|-----------|----------|
| A       | 1.60      | 55       |
| B       | 1.61      | 100      |
| C       | 1.90      | 56       |

Estos valores pueden provenir de un conjunto de datos donde las personas miden entre 1.50 y 1.95 (m) y el peso en el rango de 50 a 100 (kg). Claramente las dos variables tomarán valores que corresponden a magnitudes muy diferentes.
Ahora bien, una forma de medir similitud entre datos es utilizando la métrica de distancia euclídea (o euclidiana - ver abajo para más información). Dada la forma que toma esta métrica, los cambios en la variable Peso afectarán más al valor de la distancia euclidiana que los cambios en Altura.

### Comparación de las distancias
| Personas | Distancia |
|----------------|------------------------|
| A y B  | 45.00                   |
| B y C  | 44.00                   |
| A y C  | **1.04**                |

Observamos que, aunque las personas A y B son muy similares en altura (solo 1 cm de diferencia), la gran diferencia en peso (45 kg) hace que la distancia sea casi la misma que la de B y C, incluso cuando la diferencia de altura entre B y C es de 29 cm.
Por otro lado, A y C tienen características fisicas muy diferentes (casi el mismo peso pero con una diferencia de 30 cm en altura), y sin embargo su distancia es mucho menor que la distancia entre A y B o la de B y C.

Esto demuestra cómo una variable con una escala más grande **domina la distancia euclidiana**, haciendo que otra variable (altura) tenga un impacto mínimo.

Si normalizamos los datos, estas distancias se equilibrarán, reflejando mejor las diferencias reales entre los puntos.


## **Anexo** ¿Qué es la distancia euclidea?

La **distancia euclidiana** es una medida de la separación directa entre dos puntos en un espacio. Es la distancia más corta entre dos puntos, similar a medir con una regla.

En dos dimensiones, si tenemos dos puntos $ A(x_1, y_1) $ y $ B(x_2, y_2) $, la distancia euclidiana se calcula con la fórmula:

$$
d(A, B) = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
$$

Esta ecuación proviene del Teorema de Pitágoras, donde los cambios en $x$ e $y$ forman los catetos de un triángulo rectángulo, y la distancia entre los puntos es la hipotenusa.
