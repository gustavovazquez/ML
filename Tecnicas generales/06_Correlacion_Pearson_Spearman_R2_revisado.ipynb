# Correlación de Pearson, correlación de Spearman y $R^2$


La correlación de Pearson, el coeficiente de determinación $R^2$ y la correlación de Spearman son herramientas relacionadas, pero responden preguntas diferentes.

## 1. Correlación de Pearson

### Definición

La correlación de Pearson mide la **fuerza** y la **dirección** de la relación **lineal** entre dos variables numéricas.

Su valor se encuentra entre $-1$ y $1$:

- $r = 1$: relación lineal positiva perfecta.
- $r = -1$: relación lineal negativa perfecta.
- $r = 0$: ausencia de relación lineal.

La fórmula es:

$$
r =
\frac{
\sum_{i=1}^{n}(x_i-\bar{x})(y_i-\bar{y})
}{
\sqrt{
\sum_{i=1}^{n}(x_i-\bar{x})^2
\sum_{i=1}^{n}(y_i-\bar{y})^2
}
}
$$

### Interpretación

Pearson indica si dos variables tienden a moverse juntas siguiendo una relación aproximadamente lineal.

Por ejemplo, si $r$ es positivo y alto, valores grandes de $X$ tienden a estar asociados con valores grandes de $Y$. Si $r$ es negativo y alto en valor absoluto, valores grandes de $X$ tienden a estar asociados con valores pequeños de $Y$.

Es importante notar que Pearson **no mide causalidad**. Una correlación alta no implica que una variable cause la otra.

## 2. Coeficiente de determinación $R^2$

### Definición

El coeficiente de determinación $R^2$ mide qué proporción de la variabilidad observada en la variable dependiente $Y$ es explicada por un modelo de regresión.

Se calcula como:

$$
R^2 =
1 -
\frac{
\sum_{i=1}^{n}(y_i-\hat{y}_i)^2
}{
\sum_{i=1}^{n}(y_i-\bar{y})^2
}
$$

donde:

- $y_i$ es el valor observado.
- $\hat{y}_i$ es el valor predicho por el modelo.
- $\bar{y}$ es el promedio de los valores observados.
- $\sum (y_i-\hat{y}_i)^2$ es la suma de errores cuadrados del modelo.
- $\sum (y_i-\bar{y})^2$ es la variabilidad total de $Y$ respecto a su media.

### Interpretación

Un valor de $R^2 = 1$ indica que el modelo explica perfectamente la variabilidad de $Y$.

Un valor de $R^2 = 0$ indica que el modelo no mejora la predicción que se obtendría usando simplemente la media de $Y$.

Un valor alto de $R^2$ indica que el modelo explica una proporción importante de la variabilidad de la variable dependiente.

Aunque muchas veces se dice que $R^2$ está entre $0$ y $1$, esto es estrictamente cierto en regresión lineal ajustada por mínimos cuadrados con intercepto y evaluada sobre los mismos datos de entrenamiento. En otros contextos, especialmente al evaluar sobre datos de test, $R^2$ puede ser negativo si el modelo predice peor que usar la media de la variable respuesta.

---

## 3. Relación entre Pearson y $R^2$

Pearson y $R^2$ están relacionados, pero no significan lo mismo.

En una **regresión lineal simple** con intercepto, usando una sola variable predictora $X$, se cumple que:

$$
R^2 = r^2
$$

donde $r$ es la correlación de Pearson entre $X$ e $Y$.

Sin embargo, esta equivalencia no debe generalizarse sin cuidado. En regresión múltiple, modelos no lineales, modelos sin intercepto o evaluaciones sobre datos nuevos, $R^2$ ya no puede interpretarse simplemente como el cuadrado de la correlación de Pearson entre una variable $X$ y $Y$.

También es importante distinguir la información que aporta cada medida:

- Pearson mide la intensidad y dirección de una relación lineal entre dos variables.
- $R^2$ mide qué proporción de la variabilidad de la variable dependiente es explicada por un modelo.
- Pearson puede ser negativo o positivo.
- $R^2$ no indica dirección: al elevar $r$ al cuadrado, se pierde el signo.

## 4. Correlación de Spearman

### Definición

La correlación de Spearman mide la fuerza y dirección de una relación **monótona** entre dos variables.

Una relación monótona es aquella en la que, cuando una variable aumenta, la otra tiende a aumentar o a disminuir de forma consistente, aunque la relación no sea necesariamente lineal.

Spearman no se calcula directamente sobre los valores originales, sino sobre sus **rangos**.

En términos simples:

1. Se reemplazan los valores de $X$ por sus posiciones ordenadas o rangos.
2. Se reemplazan los valores de $Y$ por sus posiciones ordenadas o rangos.
3. Se calcula la correlación de Pearson entre esos rangos.

Por eso, Spearman suele denotarse como $\rho$ o $r_s$.

### Interpretación

Su valor también se encuentra entre $-1$ y $1$:

- $\rho = 1$: relación monótona positiva perfecta.
- $\rho = -1$: relación monótona negativa perfecta.
- $\rho = 0$: ausencia de relación monótona clara.

Spearman es especialmente útil cuando:

- La relación entre las variables es creciente o decreciente, pero no lineal.
- Hay valores atípicos que podrían afectar mucho a Pearson.
- Las variables son ordinales.
- Se quiere evaluar asociación basada en ordenamientos más que en distancias numéricas exactas.

Por ejemplo, si al aumentar $X$ también aumenta $Y$, pero siguiendo una curva, Pearson puede no ser tan alto como Spearman. En ese caso, Spearman puede capturar mejor la asociación porque se enfoca en el orden relativo de los datos.

---

## 5. Comparación resumida

| Medida | ¿Qué evalúa? | Tipo de relación | Rango | ¿Tiene signo? | Uso típico |
|---|---|---|---|---|---|
| Pearson $r$ | Asociación lineal entre dos variables | Lineal | $[-1, 1]$ | Sí | Medir correlación lineal |
| Spearman $\rho$ | Asociación monótona entre dos variables | Monótona, no necesariamente lineal | $[-1, 1]$ | Sí | Medir relación basada en rangos |
| $R^2$ | Proporción de variabilidad explicada por un modelo | Depende del modelo | Usualmente $[0, 1]$, pero puede ser negativo en test | No | Evaluar desempeño explicativo de una regresión |

---

## Idea clave

Pearson y Spearman son medidas de **asociación entre variables**.

$R^2$ es una medida de **calidad explicativa de un modelo de regresión**.

En regresión lineal simple con intercepto, $R^2$ coincide con el cuadrado de Pearson, pero conceptualmente no son lo mismo.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Generar datos sintéticos con una relación lineal aproximada
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Calcular la correlación de Pearson
r_pearson, p_pearson = pearsonr(X.flatten(), y.flatten())

# Calcular la correlación de Spearman
r_spearman, p_spearman = spearmanr(X.flatten(), y.flatten())

print(f"Correlación de Pearson:  {r_pearson:.3f}")
print(f"Correlación de Spearman: {r_spearman:.3f}")

# Entrenar un modelo de regresión lineal
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred = lin_reg.predict(X)

# Calcular el coeficiente de determinación R^2
r2 = r2_score(y, y_pred)

print(f"Coeficiente de determinación R^2: {r2:.3f}")
print(f"Pearson al cuadrado:              {r_pearson**2:.3f}")

# Visualizar la relación y la regresión
plt.figure(figsize=(7, 5))
plt.scatter(X, y, label="Datos")
plt.plot(X, y_pred, color="red", label="Recta de regresión")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Relación lineal: Pearson, Spearman y R²")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
